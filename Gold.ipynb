### Исследование суммарной концентрации всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах.

Определение функции для расчета суммарных концентрацией всех веществ и построения гистограммы.

```python
def sum_function(data,name):
    data['rougher.input.all']=data[['rougher.input.feed_ag','rougher.input.feed_pb',
        'rougher.input.feed_au','rougher.input.feed_sol']].sum(axis=1)

    data['rougher.output.all']=data[['rougher.output.concentrate_ag','rougher.output.concentrate_pb',
        'rougher.output.concentrate_au','rougher.output.concentrate_sol']].sum(axis=1)

    data['primary_cleaner.output.all']=data[['primary_cleaner.output.concentrate_ag','primary_cleaner.output.concentrate_pb',
        'primary_cleaner.output.concentrate_au','primary_cleaner.output.concentrate_sol']].sum(axis=1)

    data['final.output.all']=data[['final.output.concentrate_ag','final.output.concentrate_pb',
        'final.output.concentrate_au','final.output.concentrate_sol']].sum(axis=1)
    data[['rougher.input.all','rougher.output.all','primary_cleaner.output.all','final.output.all']].plot(
    bins=100,
    title='Распределение суммарной концентрации веществ '+name,
    kind='hist',
    grid=True,
    figsize=(9, 4),
    alpha = 0.7,
    xlim=(0,100)
    )
```

```python
sum_function(df,'на исходной общей выборке')
```

## Модель

### Функция для вычисления итоговой sMAPE


```python
def smape(true,predict):
    rougher_y=true['rougher.output.recovery']
    rougher_y_pred=predict[:,0]
    final_y=true['final.output.recovery']
    final_y_pred=predict[:,1]
    n=len(true)
    smape_rougher=((abs(rougher_y-rougher_y_pred))/((
                  abs(rougher_y)+abs(rougher_y_pred))/2)).sum()/n*100
    smape_final=((abs(final_y-final_y_pred))/((
                  abs(final_y)+abs(final_y_pred))/2)).sum()/n*100
    smape_total=smape_rougher*0.25+smape_final*0.75
    return smape_total
```

### Определение лучшей модели по метрике итоговая sMAPE

####  Поиск наилучшей модели решающего дерева с применением метода  GridSearchCV

```python
smape_scorer=make_scorer(smape,greater_is_better=False)
```

```python
%%time
print()
model = DecisionTreeRegressor(random_state=12345)
parametrs = { 'max_depth': range (2,6),
              'min_samples_leaf': range (1,8),
              'min_samples_split': range (2,10,2) }
grid = GridSearchCV(model, parametrs, cv=3,scoring=smape_scorer)
grid.fit(train_features,train_target)
print("Итоговое sMAPE лучшей модели решающего дерева на тестовой выборке:")
print(abs(grid.score(test_features,test_target).round(2)))
print()
print("Значения гиперпараметров:")
best_parameters = grid.best_estimator_.get_params()
for param_name in ['max_depth','min_samples_leaf','min_samples_split']:
    print(param_name, best_parameters[param_name])
print()
```

####  Поиск наилучшей модели случайного леса с применением метода  GridSearchCV
